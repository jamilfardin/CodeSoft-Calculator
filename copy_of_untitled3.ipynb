```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
def plot_function(f, x_range=(-10, 10), label="Function"):
    x = np.linspace(x_range[0], x_range[1], 400)
    y = [f(val) for val in x]
    plt.plot(x, y, label=label)
    plt.axhline(0, color='black',linewidth=1)
    plt.axvline(0, color='black',linewidth=1)
    plt.legend()
    plt.grid(True)
    plt.show()

# Example Usage
f = lambda x: x**2 - 4*x + 2
plot_function(f, label="x^2 - 4x + 2")
def create_table(f, x_values):
    print(f"{'x':<10}{'y':<10}")
    for x in x_values:
        y = f(x)
        print(f"{x:<10}{y:<10}")

# Example usage
x_values = np.linspace(-10, 10, 5)
create_table(f, x_values)
def shade_function(f, x_range=(-10, 10), above=True):
    x = np.linspace(x_range[0], x_range[1], 400)
    y = [f(val) for val in x]
    plt.plot(x, y, label="Function")
    plt.axhline(0, color='black', linewidth=1)
    plt.axvline(0, color='black', linewidth=1)

    if above:
        plt.fill_between(x, y, np.max(y), color='red', alpha=0.3)
    else:
        plt.fill_between(x, y, np.min(y), color='blue', alpha=0.3)

    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
shade_function(f, above=True)
def solve_system(eq1, eq2):
    x, y = sp.symbols('x y')
    sol = sp.solve([eq1, eq2], (x, y))
    return sol

# Example usage
x, y = sp.symbols('x y')
eq1 = sp.Eq(x + y, 3)
eq2 = sp.Eq(2*x - y, 1)

solution = solve_system(eq1, eq2)
print("Solution: ", solution)
def zoom_plot(f, x_range=(-10, 10), y_range=(-10, 10)):
    x = np.linspace(x_range[0], x_range[1], 400)
    y = [f(val) for val in x]
    plt.plot(x, y, label="Function")
    plt.axhline(0, color='black', linewidth=1)
    plt.axvline(0, color='black', linewidth=1)
    plt.xlim(x_range)
    plt.ylim(y_range)
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
zoom_plot(f, x_range=(-5, 5), y_range=(-10, 10))
def solve_quadratic(a, b, c):
    x = sp.symbols('x')
    eq = sp.Eq(a*x**2 + b*x + c, 0)
    solutions = sp.solve(eq, x)
    return solutions

# Example usage
solutions = solve_quadratic(1, -4, 2)
print("Quadratic solutions: ", solutions)
```
